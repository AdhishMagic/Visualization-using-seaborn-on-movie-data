# Visualization using seaborn on movie data

## Import the Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Dataset
df = pd.read_csv('/content/Trending_Movies.csv')
```

```python
# Convert release_date to datetime
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
# Check missing values
df.isnull().sum()
```

```python
sns.histplot(df['popularity'], kde=True)
plt.title("Distribution of Movie Popularity")
plt.show()
```

```python
sns.boxplot(x=df['vote_average'])
plt.title("Boxplot of Movie Ratings")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x='popularity',
    y='vote_average',
    size='vote_count',
    sizes=(20, 200)
)
plt.title("Popularity vs Rating")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x='vote_count',
    y='vote_average'
)
plt.title("Vote Count vs Rating")
plt.show()
```

```python
sns.countplot(data=df, x='original_language')
plt.title("Movies by Language")
plt.show()
```

```python
df['year'] = df['release_date'].dt.year

sns.countplot(data=df, x='year')
plt.title("Movies Released per Year")
plt.show()
```

```python
corr = df[['popularity', 'vote_average', 'vote_count']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
